```python
def male_or_female(x):
  if x == '남자':
    return 1
  elif x == '여자':
    return 0
```

```python
df['성별'].apply(male_or_female)
```

```python
df['NEW_성별'] = df['성별'].apply(male_or_female)
```

```python
# lambda 함수 : 한줄로 간단하게 쓸 수 있을 때 사용한다.
male_or_female = lambda x: 1 if x == '남자' else 0 
```

### **25. 원 핫 인코딩(One-Hot-Encoding)**

* 원 핫 인코딩은 한개의 요소는 True, 나머지 요소는 False로 만들어주는 기법
* 각각의 데이터가 관계성을 갖지 않도록 하는 기법

```python
blood_map = {'A':0, 'B':1, 'AB':2, 'O':3}
```

```python
df
```

```python
df['혈액형_code'] = df['혈액형'].map(blood_map)
```

```python
df.head()
```

```python
df['혈액형_code'].value_counts()
```

* df['혈액형_code']를 머신러닝 알고리즘에 넣어 데이터를 예측하라고 지시한다면 컴퓨터는 값들간의 관계를 스르로 형성하게 됨
* 만약 B형은 1, AB형은 2라는 값을 가지고 있을 때, 컴퓨터는 'B형 + AB형 = O형'라고 잘못된 관계를 맺을 수 있게 됨
* 별도의 column을 형성해주고 1개의 column에는 True, 나머지는 False라고 넣어줌으로 'A, B, AB, O형의 관계는 독립적이다.'라는 카테고리로 표현해주는 방식

```python
pd.get_dummies(df['혈액형_code'])
```
