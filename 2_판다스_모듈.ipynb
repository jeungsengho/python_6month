### **1-1. Series와 DataFrame**
* 1차원, 1개의 column은 Series라고 함
* 데이터프레임 : 가로축, 세로축이 있는 2차원 다양한 타입의 데이터를 저장하는 자료구조

```python
pd.Series([1, 2, 3, 4])
```

```python
a = [1, 2, 3, 4]
pd.Series(a)
```

```python
product1 = [['삼성', 990000, '갤럭시22'],
            ['애플', 300000, '맥북에어'],
            ['엘지', 1200000, '엘지그램']]
```

```python
pd.DataFrame(product1)
```

```python
df1 = pd.DataFrame(product1)
```

```python
df1
```

```python
# 컬럼 제목 변경
df1.columns = ['회사명', '가격', '상품']
```

```python
df1
```

```python
product2 = {'회사명': ['삼성', '애플', '엘지'],
            '가격': [990000, 300000, 1200000],
            '상품': ['갤럭시22', '맥북에어', '엘지그램']}
```

```python
df2 = pd.DataFrame(product2)
```

```python
df2
```
