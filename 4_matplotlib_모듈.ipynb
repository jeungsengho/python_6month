```python
plt.plot([1, 2, 3], [3, 6, 9])
plt.plot([1, 2, 3], [2, 4, 9])

plt.title('타이틀 테스트', fontsize=30)

plt.xlabel('X축', fontsize=20)
plt.ylabel('Y축', fontsize=20, rotation=0) # rotation 글자의 기울기 설정

plt.show()
```

```python
plt.plot(np.arange(10), np.arange(10)*2)
plt.plot(np.arange(10), np.arange(10)**2)
plt.plot(np.arange(10), np.log(np.arange(10)))

plt.title('범례 설정', fontsize=30)

plt.xlabel('X축', fontsize=20)
plt.ylabel('Y축', fontsize=20, rotation=0)

plt.xticks(rotation=90)
plt.yticks(rotation=30)

# plt.legend(['10*2', '10**2', 'log'], fontsize=15, loc='upper right')
plt.legend(['10*2', '10**2', 'log'], fontsize=15, loc=(0.5, 0.5), ncol=3) 
# 위치를 숫자로도 줄 수 있다.
# ncol: 범주박스의 컬럼의 개수
plt.xlim(0, 10)
plt.ylim(0.5, 20)

plt.show()
```

### **3. 스타일 세부 설정**

* [스타일세부설정](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
plt.plot(np.arange(10), np.arange(10)*2, marker='o', markersize=5, linestyle='', color='b', alpha=0.3)
plt.plot(np.arange(10), np.arange(10)**2, marker='v', markersize=10, linestyle='--', color='c', alpha=0.5)
plt.plot(np.arange(10), np.log(np.arange(10)), marker='*', markersize=15, linestyle='-.', color='r', alpha=0.7)

plt.title('마커 설정', fontsize=30)

plt.xlabel('X축', fontsize=20)
plt.ylabel('Y축', fontsize=20, rotation=0)

plt.xticks(rotation=90)
plt.yticks(rotation=30)

plt.grid()

plt.show()
```

```python
x = np.random.rand(50) # 0 ~ 1 사이의 소수
y = np.random.rand(50)

colors = np.arange(50)
area = x * y * 1000

print(x)
print(y)
print(colors)
print(area)
```

```python
# 원 그래프
plt.scatter(x, y, s=area, c=colors)
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(131)
plt.scatter(x, y, s=area, cmap='blue', alpha=0.1) # cmap='color' 모든 그래프의 색상을 동일하게 만들어준다.
plt.title('alpha=0.1')

plt.subplot(132)
plt.scatter(x, y, s=area, cmap='blue', alpha=0.5)
plt.title('alpha=0.5')

plt.subplot(133)
plt.scatter(x, y, s=area, cmap='blue', alpha=1.0)
plt.title('alpha=1.0')

plt.show()
```

```python
x = ['Math', 'Programming', 'Data science', 'art', 'English', 'Korea']
y = [67, 80, 50, 80, 10, 90]

plt.figure(figsize=(6, 3))
plt.bar(x, y, align='edge', alpha=0.6, color='red')
plt.xticks(x)
plt.ylabel('Number of Students')
plt.title('Subjects')

plt.show()
```

```python
# plt.barh() x, y축을 회전시킨다.

x = ['Math', 'Programming', 'Data science', 'art', 'English', 'Korea']
y = [67, 80, 50, 80, 10, 90]

plt.figure(figsize=(6, 3))
# plt.bar(x, y, align='edge', alpha=0.6, color='red')
plt.barh(x, y, align='center', alpha=0.7, color='pink')
plt.yticks(x)
plt.xlabel('Number of Students')
plt.title('Subjects')

plt.show()
```

```python
# 라인플롯

x = np.arange(0, 10, 0.1)
y = 1 + np.sin(x)

plt.plot(x, y)

plt.xlabel('x value', fontsize=15)
plt.ylabel('y value', fontsize=15, rotation=0)
plt.title('sin graph', fontsize=30)

plt.show()
```
